```python
import pandas as pd
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

## SentenceBERT를 이용한 문장 인코딩

https://huggingface.co/jhgan/ko-sroberta-multitask

```python
encoder = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
#한국어 sentences를 컴퓨터가 이해할 수 있는 숫자 embedding으로 표현하는 것.
embeddings = encoder.encode(sentences)

print(embeddings)
```

```python
encoder.encode('안녕하세요')
```

```python
hi1 = encoder.encode("안녕하세요")
hi2 = encoder.encode('나는 너무 슬퍼')
```

```python
# 두 문장의 유사도 측정, cosine_similarity는 2차원 벡터를 이용하므로 []로 감싸줘야됨.
cosine_similarity([hi1], [hi2])
```

## 데이터셋 로드


```python
#세브란스 병원에서 만든 정신건강 상담용 데이터셋
df = pd.read_csv('https://github.com/kairess/mental-health-chatbot/raw/master/wellness_dataset_original.csv')

df.head()
```

## 데이터 전처리

```python
#필요 없는 칼럼 제거
df = df.drop(columns=['Unnamed: 3'])

df.head()
```

```python
#챗봇의 답변이 정해지지 않은 행 제거
df = df.dropna()

print(len(df))
```

## 챗봇의 원리

```python
#0번째 행의 유저(내담자)의 발화 내용을 추출
df.loc[0, '유저']
```

```python
#추출한 발화 내용을 인코딩해서 768차원의 벡터 데이터로 변환
encoder.encode(df.loc[0, '유저'])
```

## 모든 데이터셋 인코딩

```python

df['embedding'] = pd.Series([[]] * len(df)) # dummy
#csv 파일 가장 우측에 열을 하나 추가해서 내담자의 발화를 모두 인코딩 해서 'embedding'이라는 열에 추가.
df['embedding'] = df['유저'].map(lambda x: list(encoder.encode(x)))

df.head()
```

## 예제 문장 입력

```python
text = '인생이 내 맘같지가 않아'

embedding = encoder.encode(text)

embedding
```

## 입력된 문장에 대해 유사도 계산

```python
#새롭게 입력한 문장과 csv 파일에 저장되어 있던 문장들 간의 유사도를 측정해서 가장 유사한 문장이 있는 행을 선택한다.
df['similarity'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())

df.head()
```

## 유사도가 가장 높은 답변을 채택

```python
#새롭게 입력한 문장과 csv 파일에 저장되어 있던 문장들 간의 유사도를 측정해서 가장 유사한 문장이 있는 행을 선택하고, 그 행에서의 정해진 대답을 도출한다.
answer = df.loc[df['similarity'].idxmax()]

print('구분', answer['구분'])
print('유사한 질문', answer['유저'])
print('챗봇 답변', answer['챗봇'])
print('유사도', answer['similarity'])
```

## 심리상담 챗봇 앱

```python
!pip install -q gradio
```

```python
import gradio as gr

def greet(user, history=[]):
    embedding = encoder.encode(user)

    df['distance'] = df['embedding'].map(lambda x: cosine_similarity([embedding], [x]).squeeze())
    answer = df.loc[df['distance'].idxmax()]

    history.append([user, answer['챗봇']])

    return history, history

demo = gr.Interface(fn=greet, inputs=["text", "state"], outputs=["chatbot", "state"])

demo.launch(debug=True, share=True)
```
